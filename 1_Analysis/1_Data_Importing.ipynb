```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast

movies_df = pd.read_csv(r'C:\Users\Administrator\Desktop\Studies\Python\Python_Project_Movie_Analysis\0_Dataset\movies_metadata.csv', low_memory= False)

movies_df['release_date'] = pd.to_datetime(movies_df['release_date'],errors='coerce') # Convert data type of release_date to datetime.
movies_df['genres'] = movies_df['genres'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x) # Converting the string to list.
movies_df['budget'] = pd.to_numeric(movies_df['budget'], errors='coerce') # Convert data type of budget to numeric.

def extract_country_names(country_str): # Define a function to extract country names
    try:
        countries = ast.literal_eval(country_str)
        return [country['name'] for country in countries]
    except (ValueError, SyntaxError, TypeError):
        return None

movies_df['production_countries_name'] = movies_df['production_countries'].apply(extract_country_names) # List of production countries name.

movies_df = movies_df.explode('production_countries_name')
movies_df_US = movies_df[movies_df['production_countries_name'] == "United States of America"] # Filtering all the movies that was produced in United States.

movies_df_US.head(5)
```
